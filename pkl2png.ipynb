```python
'''


'''

import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from PIL import Image

dataset = 'KAG'

if dataset == 'CAT':    
    # flist1 = glob('data/pkl_la_train/D*.pkl')
    flist1 = glob('data/pkl_la_test/D*.pkl')
    str_train_or_test = 'test'

elif dataset == 'KAG':
    flist1 = glob('data/KAG_2ndAnnual/pkl_la_validate/*.pkl')
    str_train_or_test = 'train'

# print(flist1)
display_img = False

for j, fname in enumerate(flist1):
    if True:
        
        if dataset=='CAT':
            ID = os.path.split(fname)[1][:10]
        elif dataset =='KAG':
            ID = os.path.split(fname)[1][:-17]
        
        print(j, fname, ID[:])
                
        with open(fname, 'rb') as f:
            dat = pickle.load(f)
        la_img_list = dat['la_img_list']
        est_la_img_list = dat['est_la_img_list']
        nLA = len(la_img_list)
        #print(len(la_img_list), len(est_la_img_list))

        ''' save as .png file '''
        if dataset=='CAT':
            dir1 = os.path.join('data/CAT', str_train_or_test, ID, 'oriLA')
        elif dataset =='KAG':
            dir1 = os.path.join('data/KAG_2ndAnnual/png', str_train_or_test, ID, 'oriLA')
        
        os.makedirs(dir1, exist_ok=True)
        for k in range(nLA):
            img1 = la_img_list[k]
            img1 = img1[40:-40, 40:-40]
            img1 = 255*img1/np.max(img1)
            img1 = img1.astype(np.uint8)
            im = Image.fromarray(img1)
            im.save(os.path.join(dir1, ID+'_oLA{k:02d}.png'.format(k=k)))
        
        if dataset=='CAT':
            dir2 = os.path.join('data/CAT', str_train_or_test, ID, 'estLA')
        elif dataset=='KAG':
            dir2 = os.path.join('data/KAG_2ndAnnual/png', str_train_or_test, ID, 'estLA')
            
        os.makedirs(dir2, exist_ok=True)
        for k in range(nLA):
            img1 = est_la_img_list[k][5]
            img1 = img1[:, :]
            img1 = 255*img1/np.max(img1)
            img1 = img1.astype(np.uint8)
            im = Image.fromarray(img1)
            im.save(os.path.join(dir2, ID+'_eLA{k:02d}.png'.format(k=k)))        
                
        if display_img:
            plt.figure(f'file = {fname}', figsize=(12, 5))
            for k in range(nLA):
                plt.subplot(2, nLA, k+1)
                la_img = la_img_list[k]
                plt.imshow(la_img, cmap='gray', clim=(0, 0.7*np.max(la_img)))

                plt.subplot(2, nLA, nLA+k+1)
                la_img_e = est_la_img_list[k][5]
                plt.imshow(la_img_e, cmap='gray', clim=(0, 0.7*np.max(la_img_e)))
        
```
